## Part 4 - Visualization - Correlation Heatmap

### Task 4.1

Create a correlation dataframe for the dataset (all columns against all columns).

Use seaborn to plot 2 correlation heatmaps:
1. Where values are going from -1 to 1.
2. Where values are going from 0 to 1.

````python
corr_matrix = df.corr() # calc correlations

# heatmap - values from -1 to 1 (full range)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap (-1 to 1)")
plt.show()

# heatmap - values from 0 to 1 (positive correlations only)
positive_corr = corr_matrix.where(corr_matrix > 0, 0)
plt.figure(figsize=(8, 6))
sns.heatmap(positive_corr, annot=True, cmap='Blues', vmin=0, vmax=1)
plt.title("Correlation Heatmap (0 to 1)")
plt.show()
````

### Task 4.2

Look at the heatmap, choose the pair of columns with the highest correlation between them,
and draw graph to show the linear relation between those 2 columns.
You have to choose the correct graph type to show that!

````python
col1, col2 = "residual sugar", "density" # HARDCODED - decided by looking at the heatmap, 0.84 is the highest... 
print(f"The pair of columns with the highest correlation is: {col1} and {col2}")

sns.regplot(x=df[col1], y=df[col2], line_kws={"color": "red"}) # plot the density / residual sugar samples along with its linear regression line
plt.title(f"Linear regression between {col1} and {col2}")
plt.xlabel(col1)
plt.ylabel(col2)
plt.show()
````
